## Ejemplo Plomero

```python
# --- SISTEMA IA: GESTIÓN DE PLOMERÍA INTERACTIVA ---

# Iniciamos con la lista de plomeros vacía y un diccionario de trabajo vacío
plomero_actual= []
losplomeros = []

print("==============================================")
print("BIENVENIDO AL SISTEMA DE IA 'PLOMERO-TECH 2026'")
print("==============================================")

# El ciclo while True mantiene el programa funcionando hasta que el usuario decida salir
while True:
    print("\n--- MENÚ PRINCIPAL ---")
    print("1. Registrar nuevo plomero (IA Database)")
    print("2. Consultar recomendación de IA para un cliente")
    print("3. Ver lista completa de plomeros disponibles")
    print("4. Apagar sistema")

    opcion = input("\nSeleccione una opción (1-4): ")

    # OPCIÓN 1: AGREGAR DATOS (Lista de diccionarios)
    if opcion == "1":
        print("\n--- REGISTRO DE TALENTO ---")
        # Reiniciamos el diccionario vacío para el nuevo registro
        plomero_actual = {}

        # Solicitamos datos al usuario
        nombre = input("Nombre del plomero: ")
        especialidad = input("Especialidad (fugas/drenaje/instalacion): ").lower()

        # Guardamos los datos en el diccionario utilizando claves
        plomero_actual["nombre"] = nombre
        plomero_actual["especialidad"] = especialidad

        # Agregamos el diccionario a la lista global
        losplomeros.append(plomero_actual) # Corregido el nombre de la variable de 'los_plomeros' a 'losplomeros'
        print(f"¡Sistema IA: Plomero {nombre} registrado exitosamente!")

    # OPCIÓN 2: SIMULACIÓN DE IA (Lógica de decisión)
    elif opcion == "2":
        print("\n--- ASISTENTE DE DIAGNÓSTICO IA ---")
        if len(losplomeros) == 0:
            print("IA: No hay plomeros en la base de datos para recomendar.")
        else:
            problema = input("IA: ¿Qué problema tiene el cliente? (fugas/drenaje/instalacion): ").lower()
            encontrado = False

            # El ciclo for recorre cada diccionario dentro de la lista
            for plomero in losplomeros:
                # La IA compara el problema del cliente con la especialidad del plomero
                if plomero["especialidad"] == problema:
                    print(f"\nIA RECOMIENDA: Para un problema de '{problema}', el mejor es: {plomero['nombre']}")
                    encontrado = True

            if encontrado == False:
                print("IA: Lo siento, no tengo un especialista para ese problema en mi base de datos.")

    # OPCIÓN 3: RECORRIDO DE DATOS (Ciclo for)
    elif opcion == "3":
        print("\n--- BASE DE DATOS DE PLOMEROS ---")
        if len(losplomeros) == 0:
            print("La lista está vacía.")
        else:
            # Utilizamos for para iterar y mostrar cada diccionario almacenado
            for p in losplomeros:
                print(f"Nombre: {p['nombre']} | Especialidad: {p['especialidad']}")

    # OPCIÓN 4: SALIR
    elif opcion == "4":
        print("\nIA: Cerrando protocolos de seguridad... ¡Adiós!")
        break # Rompe el ciclo while True para finalizar el programa

    # MANEJO DE ERRORES DE MENÚ
    else:
        print("\n[Error]: Opción no válida. Intente de nuevo.")

print("\n--- PROGRAMA FINALIZADO ---")
print("Programa realizado por Iris Chavez NL = 11")
```
