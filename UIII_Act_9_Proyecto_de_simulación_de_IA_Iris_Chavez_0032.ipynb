```python
# --- PLOMERO-BOT V3.0: ASISTENTE LÓGICO DE REPARACIONES ---

# Definición de la lista vacía para almacenar el historial de problemas
historial_averias = []

print("************************************************")
print("* BIENVENIDO A PLOMERO-BOT: TU IA DE FONTANERÍA *")
print("************************************************")

sistema_activo = "si"

# El ciclo while mantiene el programa funcionando interactivamente
while sistema_activo == "si":
    print("\n--- MENÚ PRINCIPAL ---")
    print("1. Iniciar diagnóstico inteligente")
    print("2. Ver historial de problemas reportados")
    print("3. Calcular presupuesto estimado")
    print("4. Salir del sistema")

    opcion = input("\nSeleccione una opción (1-4): ")

    # Estructura elif para manejar las opciones del menú
    if opcion == "1":
        print("\n--- MÓDULO DE DIAGNÓSTICO IA ---")
        ubicacion = input("¿El problema es en la 'cocina' o el 'baño'? ").lower()
        problema = input("¿Es una 'fuga' o un 'tapon'? ").lower()

        # Lógica de IA basada en reglas anidadas
        if ubicacion == "cocina":
            if problema == "fuga":
                diagnostico = "Fuga en el fregadero. Sugerencia: Ajustar la tuerca del sifón."
            else:
                diagnostico = "Obstrucción en tarja. Sugerencia: Usar desatascador líquido."
        elif ubicacion == "baño":
            if problema == "fuga":
                diagnostico = "Fuga en sanitario. Sugerencia: Revisar la válvula de llenado."
            else:
                diagnostico = "Inodoro tapado. Sugerencia: Utilizar bomba de vacío (sopapa)."
        else:
            diagnostico = "Ubicación desconocida. Análisis fuera de rango."

        print(f"IA: {diagnostico}")

        # Agregamos el diagnóstico a la lista de historial
        historial_averias.append(diagnostico)
        print("¡Diagnóstico guardado en la lista!")

    elif opcion == "2":
        print("\n--- HISTORIAL DE REPORTES ---")
        # Verificamos si la lista está vacía con un condicional
        if len(historial_averias) == 0:
            print("El historial está vacío actualmente.")
        else:
            # Para imprimir la lista sin ciclo 'for', usamos un 'while' con índice
            indice = 0
            while indice < len(historial_averias):
                print(f"- {historial_averias[indice]}")
                indice = indice + 1 # Incremento manual del índice

    elif opcion == "3":
        print("\n--- CALCULADORA DE COSTOS ---")
        costo_base = 150 # Operadores aritméticos
        horas = int(input("¿Cuántas horas tomaría la reparación? "))

        # Cálculo de presupuesto simple
        total = costo_base + (horas * 80)
        print(f"El presupuesto estimado es de: ${total} pesos.")

    elif opcion == "4":
        print("\nIA: Finalizando protocolos... ¡Gracias por usar Plomero-Bot!")
        sistema_activo = "no" # Condición para romper el ciclo while

    else:
        print("\nError: Opción no válida. Intente de nuevo.")

print("\n--- PROGRAMA FINALIZADO ---")
print("Programa realizado por: Iris Chavez")

```
